```python
import os
import json
directory = os.getcwd()
lyrics = []
for filename in os.listdir(directory):
    if filename.endswith(".json"):
        with open(filename) as json_file:  
            data = json.load(json_file)
        lyrics.append(data['songs'][0]['lyrics'])
        continue
    else:
        continue
```
